```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import re

print('--------')
print('--------')
print('Extra Credit #1 - due: 3/25')
print('')
print('  by David Morgan')
print('')
print('    Scrape job skill keywords using Beautiful Soup and Regular Expressions.')

url1 = 'https://www.simplyhired.com/job/zMJIysVrwLVHFGtPy74Gh3CXrckMx-M_Ps-gnvgqeWgWuVw4S_5_LQ?isp=0&q=data+analyst'
url2 = 'https://www.simplyhired.com/job/L8sLPMdPV7LGhQ3hTKMjN3fxPEKt3NbVrvz2xdgdO65KidxaYOIgJw?isp=0&q=data+analyst'
url3 = 'https://www.simplyhired.com/job/ICO2n7XHI-9aI40jea5ZGn4EQEQB7GprZ0yGsLiEfKsaIG6jdoPYFA?isp=1&sjdu=s31xlD7RXxHYPSSyqVtwh_6lh4JToAM7s9WvRzd5wC3Q5yRWr8j-_0hcWh40xuWla-9SvgRMikLKBXH1h5wAHcrXe4UYSes0eYUcxg&q=data+analyst'
url4 = 'https://www.simplyhired.com/job/y2vjasdP5_COssbO2AJHRO4wEHHuZNfsWNJPAUHCxLCYiR-6iNdmPg?isp=0&q=data+analyst'

urlList = [url1, url2, url3, url4]

for siteCount, url in enumerate(urlList):

    print('')
    print('--------')
    print('--------')
    print('')
    print('  URL #', siteCount+1, ':', url)
    print('')
    
    htmlfile=urlopen(url)

    soup = bs(htmlfile,'html.parser')

    print('Page title: ', soup.title.string.strip())
    print('')
    print('----')

    skillsFound = soup.find(class_="viewjob-entities")
    
    print('')
    print('HTML Skill List [div contents]: ')
    print('')    
    
    for i, tag in enumerate(skillsFound):

        if i == 1:

            print(tag.contents)
            skillsTagged = tag.contents
            
    print('')
    print('----')            
    print('')
    
    print('**** SKILLS -- Text extracted with RegEx: ')
    print('')
    
    # example HTML
    # [<li class="nav-item">SQL</li>, <li class="nav-item">Python</li>]
       
    try: 
        
        for skill in skillsTagged:    

            skillString = str(skill)

            regex_ToFind = '>(.+?)<'
            pattern = re.compile(regex_ToFind)             

            skillClean = re.findall(pattern,skillString)
            
            print('    ', *skillClean)

    except:
        print('')
        print('    !!! -- Error with RegEx -- !!!')
    
    print('')
    print('----')
    print('')
    print('Full Job Description: ')
    print('')

    jobDesc = soup.find(class_="viewjob-description")

    jobDesc2 = jobDesc.text.strip()

    print(jobDesc2)
    print('')
    print('--------')
```
